```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_data=pd.read_csv("C:/Users/vishw/Downloads/archive (2)/creditcard.csv")
```

```python
credit_data.head()
```

```python
credit_data.info() 
```

```python
credit_data.isnull().sum()
```

```python
credit_data['Class'].value_counts()
```

```python
normal=credit_data[credit_data.Class==0]
fraud=credit_data[credit_data.Class==1]
```

```python
print(normal.shape)
print(fraud.shape)
```

Undersampling

```python
normal_sample=normal.sample(n=492)
```

```python
new_data=pd.concat([normal_sample,fraud], axis=0)
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
X=new_data.drop('Class', axis=1)  #features and target
y=new_data['Class']
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Logistic Regression

```python
model=LogisticRegression()
model.fit(X_train, y_train)

X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction, y_train)
print('Accuracy on training data:' ,training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction, y_test)
print('Accuracy score on test data:', test_data_accuracy)
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_train,X_train_prediction)
print("classification Report:")
print(report)
```
